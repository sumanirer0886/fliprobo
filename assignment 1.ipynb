```python
#9) Write a python program to scrape mentioned details from dineout.co.in :
       # i) Restaurant name
        #ii) Cuisine
       # iii) Location 
       # iv) Ratings
       # v) Image URL
```

```python
src=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
sup=BeautifulSoup(src.content)
```

```python
ht=[]
for i in sup.find_all('a',class_="restnt-name ellipsis"):
    ht.append(i.text)
```

```python
rating=[]
for i in sup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
```

```python
cuisine=[]
for i in sup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text)
```

```python
fcui=[]
for i in cuisine:
    fcui.append(i.split('|')[1])
```

```python
loc=[]
for i in sup.find_all('div',class_="restnt-loc ellipsis"):
    loc.append(i.text)
```

```python
pc=[]
for i in sup.find_all('img',class_="no-img"):
    pc.append(i.get('data-src'))
```

```python
import pandas as pd
hoteld=pd.DataFrame({'location':loc,'coisine':fcui,'rating':rating,'hotel name':ht,'images':pc})
```
